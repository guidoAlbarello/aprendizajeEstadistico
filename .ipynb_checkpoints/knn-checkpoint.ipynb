```python
import matplotlib.pyplot as plt
import math as math
import pandas as pd
import numpy as np
```

```python
def uniformRect(h,w,a):
    x = np.random.uniform(-w/2, w/2,a)
    y = np.random.uniform(-h/2, h/2,a)
    return list(zip(x,y, ['black'] * a))
```

```python
def d(x,y,c):
    return [math.sqrt(math.pow(x[0] - y[0],2) + math.pow(x[1] - y[1],2)),0,c]
```

```python
def voteColor(v):
    green = 0 
    red = 0
    for x in v:
        if x == 'green':
            green = green +1
        else:
            red = red + 1
    return 'green' if green >= red else 'red'
```

```python
def knn(p,s,k):
        for a in range(len(p)):
            neighbours = []
            distances = []
            for y in s: 
                dist = d((p[a][0], p[a][1]), (y[0], y[1]), y[2])
                distances.append(dist)
            
            for j in range(k):
                m = 9999999999999999999999999.9
                idx = 0
                for i in range(len(distances)):
                    if distances[i][1] == 0:
                        if distances[i][0] < m: 
                            m = distances[i][0]
                            idx = i
                distances[idx][1] = 1
                neighbours.append(distances[idx][2])
            temp = list(p[a])
            temp[2] = voteColor(neighbours)
            p[a] = tuple(temp)
        return p
```

```python
%matplotlib inline
n = 300
b = 300
k = 10
points = uniformRect(8,8, n)

mean1 =[-1,0]
cov1 = [[1,0],[0,1]]
mean2 = [1,0]
cov2 = [[1,0],[0,1]]


n1 = np.random.multivariate_normal(mean1, cov1, int(b/2)).T
plt.plot(n1[0], n1[1],'x')

n2 = np.random.multivariate_normal(mean2, cov2, int(b/2)).T
plt.plot(n2[0], n2[1], 'x')

s = list(zip(n1[0],n1[1], ['red'] * len(n1[0])))+list(zip(n2[0],n2[1], ['green'] * len(n2[0])))
points = knn(points, s, k)


plt.scatter([i[0] for i in points], [i[1] for i in points], c = [i[2] for i in points], s=3*3.14)

plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```
